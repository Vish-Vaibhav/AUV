```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

img = cv.imread("noa.jpg")
img2 = cv.imread("noa.jpg")
# cv.imshow('name',img)
# cv.waitKey(1000)

print(img.shape)
plt.imshow(img)
plt.show()
```

```python
scale_factor = 0.3
new_img = cv.resize(img, None, fx = scale_factor, fy = scale_factor, interpolation = cv.INTER_AREA)
plt.imshow(new_img)
plt.show()
```

```python
print(new_img.shape)
```

```python
robo = new_img[ 0:300, 100:270]
img[650:950, 60:230] = robo
img[650:950, 930:1100] = robo
```
